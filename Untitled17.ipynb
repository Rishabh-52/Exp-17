```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}


df = pd.DataFrame(data)


print(df)
```

```python
#1.Line Chart(Trend)

plt.plot(df['Days'], df['Study_Hours'], marker='o', linestyle='-')


plt.title('Rishabh ka chat')
plt.xlabel('Days')
plt.ylabel('Study Hours')


plt.show()
```

```python
#Advance Line Chart
plt.figure(figsize=(7,4))
plt.plot(df['Days'], df['Study_Hours'], marker='o',color='green', label='Study Hours')
plt.plot(df['Days'], df['Marks'], marker='s',color='red', label='Marks')


plt.xlabel("Days")
plt.ylabel("Values")
plt.title("Rishabh ka advanced chat")


plt.legend()
plt.show()
```

```python
from matplotlib.lines import lineStyles
#2. Bar Chart(Comparison)

plt.bar(df['Days'],df['Marks'],color='lime')
plt.title("Rishabh")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.grid(axis='y',linestyle='--',alpha=1)


plt.show()
```

```python
plt.bar(df['Days'],df['Marks'],color='orange')
plt.title("Rishabh")
plt.xlabel("Days")
plt.ylabel("Marks")


plt.show()
```

```python
#Advanced Bar Graph
bars=plt.bar(df['Days'],df['Marks'],color='yellow',edgecolor='violet')
for bar in bars:
  height=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,height,str(height),ha='center',va='bottom')


plt.title("Rishabh")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.grid(axis='y')


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(len(df['Days']))
width=0.35
plt.bar(x-width/2,df['Study_Hours'],width,label='Study Hours',color='Red',edgecolor='Black')
plt.bar(x+width/2,df['Marks'],width,label='Marks',color='Yellow',edgecolor='violet')


plt.xticks(x,df['Days'])


plt.title('Study Hours vs Marks')
plt.xlabel('Days')
plt.ylabel('Time/Marks')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#Example data(more values for better histogram)
marks=[50,60,55,70,65,75.80,85,90,45,68,72,77,82,88]
#Calculate mean
mean_value=np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks,bins=3,color='maroon',edgecolor='#FFDB58')



#Mean Line
plt.axvline(mean_value,color='red',linestyle='dashed',linewidth=2,label='Mean')
plt.title('Rishabh')
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.5)



plt.show()
```

```python
#4. Scatter Plot(Relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'])
plt.title('Rishabh')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()

```

```python
#Condtional scatter plot
#Example category
df["Result"]=["Pass","Fail","Pass","Pass","Fail"]
colors=['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.title('Rishabh')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Dataset 2
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North',
               'South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}
df = pd.DataFrame(data)
print(df)
```

```python
#Line Plot(Dataset 2)
sns.lineplot(x='Day',y='Sales',data=df,errorbar=None)
plt.title('Rishabh')

plt.show()
```

```python
#Bar Plot(Dataset 2)
sns.barplot(x='Day',y='Sales',data=df,errorbar=None)
plt.title('Rishabh')

plt.show()
```

```python
#Histogram
sns.histplot(df['Sales'],bins=5)
plt.title('Rishabh')

plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x='Sales',y='Profit',data=df)
plt.title('Rishabh')

plt.show()
```
